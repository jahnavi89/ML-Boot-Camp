```python
movies_title=pd.read_csv('u.item',sep="\|",header=None)
```

```python
movies_title.shape
```

```python
movies_titles=movies_title[[0,1]]
movies_titles.columns=["item_id","title"]
movies_titles.head()
```

```python
df=pd.merge(df,movies_titles,on="item_id")
```

```python
df.head()
```

```python
ratings=pd.DataFrame(df.groupby('title').mean()['rating'])
```

```python
ratings.head()
```

```python
ratings['num of ratings']=pd.DataFrame(df.groupby('title').count()['rating'])
```

Creating Recommender System

```python
df.head()
```

```python
moviemat=df.pivot_table(index="user_id",columns="title",values="rating")
```

```python
moviemat.head()
```

```python
angrymen_user_ratings=moviemat['12 Angry Men (1957)']
```

```python
angrymen_user_ratings.head()
```

```python
similar_to_angrymen=moviemat.corrwith(angrymen_user_ratings)
```

```python
similar_to_angrymen
```

```python
corr_angrymen=pd.DataFrame(similar_to_angrymen,columns=['correlation'])
```

```python
corr_angrymen.dropna(inplace=True)
```

```python
corr_angrymen.head()
```

```python
corr_angrymen.sort_values('correlation',ascending=False).head(10)
```

```python
ratings
```

```python
corr_angrymen=corr_angrymen.join(ratings['num of ratings'])
```

```python
corr_angrymen.head()
```

```python
corr_angrymen[corr_angrymen['num of ratings']>100].sort_values('correlation',ascending=False)
```

To recomend any movie

```python
def predict_movies(movie_name):
    movie_user_ratings=moviemat[movie_name]
    similar_to_movie=moviemat.corrwith(movie_user_ratings)
    corr_movie=pd.DataFrame(similar_to_movie,columns=['correlation'])
    corr_movie.dropna(inplace=True)
    corr_movie=corr_movie.join(ratings['num of ratings'])
    
    predictions=corr_movie[corr_movie['num of ratings']>100].sort_values('correlation',ascending=False)
    
    return predictions
```

```python
predict_my_movie=predict_movies("Titanic (1997)")
```
